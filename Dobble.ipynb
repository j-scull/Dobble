```python
import random
```

```python
import emoji
imageDict = dict()
fin = open('C:/Users/User/Desktop/emoji_names.txt',"r")
lines = fin.readlines()
for i, el in enumerate(lines):
    imageDict[i+1] = emoji.emojize(el.strip())
    print(i+1,imageDict[i+1],end=' ')
```

```python
def create_deck(nIm, show = False):
    """
    Creates a dictionary with card number as key and a set of nIm emojis as value
    show: if true prints out the dictionary
    Returns: deck, a dictionary
    nIm the number of images per card
    """
    # nIm - 1 must be prime
    # Cards must have 3, 4, 6 or 8 images
    
    if nIm > 8:
        print("Number of image can not be greater than 8")
        return
    
    deck = {}
    n = nIm - 1
    r = range(n)
    rp1 = range(n+1)
    c = 0

    # First card
    c += 1
    if show == True:
        print('Card %2d: ' % c, end='');

    for i in rp1:
        try:
            deck[c].add(imageDict[i+1])
        except KeyError as error:
            deck[c] = {imageDict[i+1]}
    
    if show == True:
        print(deck[c])


    # n following cards
    for j in r:
        c = c+1
        deck[c] = {imageDict[1]}
        for k in r:
            deck[c].add(imageDict[n+2 + n*j +k])
        if show == True:
            print('Card %2d: ' % c, end='');
            print(deck[c])

    # n x n following cards
    for i in r:
        for j in r:
            c = c+1
            deck[c] = {imageDict[i+2]}
            for k in r:
                deck[c].add(imageDict[(n+1 +n*k + (i*k+j) % n)+1])
            if show == True:
                print('Card %2d: ' % c, end='');
                print(deck[c])
    
    return deck
```

```python
def check_validity(d, verbose=False):
    """
    Iterates through a deck, compares each card to every other card once
    Returns False if the number of common images in any two cards is not equal to one
    Otherwise returns True
    d: a dictionary representing a dobble deck
    """
    check = True
    for l in range(1, len(d) ):
        for k in range(l+1, len(d)+1):
            if verbose == True:
                print('Card', l, 'and Card', k, ':', d[l].intersection(d[k]))
            if len(d[l].intersection(d[k])) != 1:
                check = False
    return check        
    
```

```python
class DobbleCard (object):
    
    def __init__(self, c):
        """
        Creates a DobbleCard object
        c: a set of emoji's
        """
        self.card = c
        
    def get_card(self):
        """
        returns the set of emoji's that make up an individual card
        """
        return self.card
```

```python
class DobbleDeck (object):
    
    
    def __init__ (self, d):
        """
        Creates self.deck, a deck of DobbleCard objects
        d: a dictionary containing sets of emoji images
        """
        self.deck = []
        for c in d:
            card = DobbleCard(d[c])
            self.deck.append(card)
        
    def add_card(self, c):
        """
        Adds a card to self.deck
        c: a DobbleCard object
        """
        self.deck.append(c)
    
    def remove_card(self):
        """
        Removes a card from self.deck 
        Returns a Dobblecard object
        """
        p = self.deck.pop(0)
        return p   
    
    def shuffle(self):
        """
        Shuffles self.deck
        """
        for i in range(len(self.deck)-1, 0, -1):
            r = random.randint(0, len(self.deck) - (1 + i))
            self.deck[i], self.deck[r] = self.deck[r], self.deck[i]
            
    def play_cards(self):
        """
        Creates copies of the first two cards as lists
        Prints these beside each other
        """
        c1 = [c for c in self.deck[0].get_card()]
        c2 = [g for g in self.deck[1].get_card()]
        for j in range(27):
            if j%9 < 3:
                try:
                    print(c1[(j%3 + 3*(j//9))], end="")
                except IndexError as error:
                    print("  ", end="")
            elif 6 <= j%9 < 9:
                try:
                    if j%9 == 8:
                        print(c2[(j%3 + 3*(j//9))])
                    else:
                        print(c2[(j%3 + 3*(j//9))], end="")
                except IndexError as error:
                    print("  ", end="")
            else:
                print("  ", end="")
        print() 
```

```python
def dobbleGame(d):
    """
    Creates a DobbleDeck object 
    Prompts user for number of cards to be used in each game   
    Plays a game of dobble:
    Show the first two cards in the deck, removes the first card, repeats for "number of cards" entered
    Returns all removed cards to the deck
    Repeats until '0' is entered.
    
    d: a dictionary representing dobble cards
    assumes d is a valid dobble deck
    """
   
    
    print("===============PLAY DOBBLE===============")
    print()

    dobble = DobbleDeck(d)
    maxCards = len(d) -1
    
    
    while True > 0:
        
        try:
            
            num = int(input("How many cards (<" + str(maxCards) + ")? Enter '0' to exit  "))
            if num == 0:
                break
            elif num > maxCards or num < 0:
                print("Number of cards must be between 0 and",  maxCards)
            else:
                
                dobble.shuffle()
                usedCards = []
                aScore = 0
                bScore = 0
                print("Enter 'd' to record a draw")
                
                while num > 0:
            
                    #show two cards
                    dobble.play_cards()
                    win = str(input("Who wins (A or B)? "))
                    
                    # prompt user again if incorrect letter entered
                    while win.lower() != 'a' and win.lower() != 'b' and win.lower() != 'd':
                        win = str(input("Who wins (A or B)? "))
                        
                    if win.lower() == 'a':
                        aScore += 1
                    elif win.lower() == 'b':
                        bScore += 1
                    
                    #remove the first card
                    used = dobble.remove_card()
                    usedCards.append(used)
                    num -= 1
                
                print('Score:')
                print('A:', aScore)
                print('B:', bScore)
                #return used cards to the deck
                for u in usedCards:
                    dobble.add_card(u)
               
        except ValueError as error:
            print('Number of cards must be between 0 and', maxCards)

    print('Finished!')
```
